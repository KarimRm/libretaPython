```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Descarga los datos directamente con pandas, muestra el tipo de datos m√°s importantes
csv_url = "/content/data/Nacimientos_2023.csv"

df_nacimientos = pd.read_csv(csv_url, engine='python')

df_nacimientos
```

```python
#Indice con las fechas, usando una transformaci√≥n a tipo timestamp.
df_nacimientos['FECHANACIMIENTO'] = pd.to_datetime(df_nacimientos['FECHANACIMIENTO'], format='%d/%m/%Y')

df_nacimientos_by_fecha = df_nacimientos.set_index('FECHANACIMIENTO')

df_nacimientos_by_fecha
```

```python
df_nacimientos_by_fecha.info()
```

```python
#Filtra los datos para un solo a√±o y para un estado solamente, en este caso Sonora(su numero de entidad es 26)

df_nacimientos_byYear = df_nacimientos_by_fecha.loc['2023-01-01':'2023-01-31']
df_nacimientos_byYear_byState = df_nacimientos_byYear.loc[df_nacimientos_byYear['ENTIDADNACIMIENTO'] == 26]
df_nacimientos_byYear_byState
```

```python
#imprimir la columna sexo
df_nacimientos_byYear_byState['SEXO']
```

```python
#Modifica los valores de variables cualitativas para que sean m√°s representativas en su an√°lisis

df_nacimientos_byYear_byState['SEXO'] = df_nacimientos_byYear_byState['SEXO'].replace({1 : 'Masculino', 2: 'Femenino'})

df_nacimientos_byYear_byState['SEXO']
```

```python
#Realiza un an√°lisis exploratorio b√°sico usando pandas, haciendo algunas gr√°ficas que muestren algunos detalles importantes o interesantes y con algunas regrupaciones simples o complejas.

df_MyH = df_nacimientos_byYear_byState.filter(like='SEXO').value_counts()

df_MyH.plot(kind='pie').set_title('Porcentaje de nacimientos por sexo a√±o 2023')
```

```python
#Cambiamos los valores de la columna 'SECONSIDERAINDIGENA' a tipo texto para poder presentar mejor los datos
df_nacimientos_byYear_byState['SECONSIDERAINDIGENA'] = df_nacimientos_byYear_byState['SECONSIDERAINDIGENA'].replace({1 : 'Si', 2: 'No'})
```

```python
df_nacimientos_ind = df_nacimientos_byYear_byState[['EDAD','MUNICIPIONACIMIENTO']]
df_nacimientos_ind
```

```python
df_municipal = df_nacimientos_ind.loc[df_nacimientos_ind['MUNICIPIONACIMIENTO'] <= 72]
df_municipal.plot(x='MUNICIPIONACIMIENTO', y='EDAD', kind='scatter').set_title('Nacimientos por edad y municipio')
```
